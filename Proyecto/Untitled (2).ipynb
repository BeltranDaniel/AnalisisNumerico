# Proyecto de Análisis numérico

En el siguiente ejemplo se observa una simulacion del modelo empleado tomando la condicion inicial T(0) = 0

```r
#install.packages("deSolve")
library(deSolve)
# La función ode() requiere obligatoriamente, varias cosas que debemos agregar
# ode(valores iniciales, tis , func, parms, method, ...)
R = 2912
Q = 342
a = 0.30
o = 5.67e-8

fp = function(t,y, parms){
  s = (1/R)*(Q*(1-a)-o*y^4)
  return(list(s)) # ode requiere salida sea una lista
}
tis= seq(0,10000,1)
# Usamos la función ode()
sol = ode(c(0), tis, fp, parms=NULL, method = "rk4") # método Runge Kutta orden 4
# Salida
tabla = cbind(tis, sol[,2] )
colnames(tabla) = c("ti", " Ti ")
tabla
# Representación
plot(tis, sol[,2] )
```
